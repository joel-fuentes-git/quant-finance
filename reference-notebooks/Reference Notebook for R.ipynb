### Define a function

```r
mydchisq = function(x, df = 1) {
  fx =  x^(df/2-1)* exp( -x/2 ) / ( gamma(df/2) * 2^(df/2) )
  return(fx)  
}

# always the last line will be returned
mydchisq2 = function(x, df = 1) {
  x^(df/2-1)* exp( -x/2 ) / ( gamma(df/2) * 2^(df/2) )
}

mydchisq(3)
mydchisq(3,1)
dchisq(3,1)
```

### Basic Plot
#### for detailed option, ?par
#### e.g. type='l', col = "pink", cex = 0.5, pch = "#", etc ..

```r
x = seq(-3,3,length = 100)
y = dnorm(x, mean = 0, sd = 1)
z = y + rnorm(100, mean = 0, sd = 0.05)
w = y + rnorm(100, 0, 0.01)

par(mfrow=c(1,2)) # separate the graphic window 
plot(x, z, pch="*", col=2, main="plot1")
points(x, w, pch="o", col=3)  # add points
lines(x, y, col=4) # add a line
abline(v = 0, lty=2 , col=6) # add a straight line

plot(x, y, xlab = "x", ylab="f(x)", pch="^", main="Density plot of Normal Distribution")
dev.off() # kill the graphic, in the same time the effect of 'par' will be disappeared
```

### Regression and diagnostic plots

```r
data(cars)
attach(cars) ## we can directly access the name of variables in 'cars' data set
distL = log(dist)

fit = lm( speed ~ distL )
fit
summary(fit)
residuals(fit)
fitted(fit)

library(MASS)  ## Standardized residual
stdres(fit)
       
par(mfrow=c(2,2))  ## divide the graphic window
plot(fit)

par(mfrow=c(2,2))
plot(distL, speed, main="Scatter plot")
abline(fit, col=2)
plot(fitted(fit), residuals(fit), main="Residual Plot")
abline(0,0, lty=2, col=2)
plot(fitted(fit), stdres(fit), main="Standardized residual plot")
abline(0, 0, lty=2, col=2)
qqnorm(stdres(fit))
abline(0, 1, lty=2, col=2)
```
